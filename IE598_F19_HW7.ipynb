```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.model_selection import train_test_split

data = pd.read_csv(r"C:\Users\jazoi\Downloads\ccdefault.csv")
data.head()
```

```python
#PART ONE
import time

y = data.iloc[:,24]
X = data.iloc[:, 1:-1]

start = time.clock()

clf = RandomForestClassifier(n_estimators = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 1)

cv_scores_train = cross_val_score(clf, X_train, y_train, cv=10)
cv_scores_test = cross_val_score(clf, X_test, y_test, cv=10)

CV_accuracyScore_DF = pd.DataFrame(
    {
        'CV Fold': range(10),
        'In-sample Score': cv_scores_train,
    })

end = time.clock()
print('Amount of time to run: ', end - start, 'seconds')


CV_accuracyScore_DF.head(10)
```

```python
# PART TWO
# Assistance from class text, DataCamp, and https://www.programcreek.com/python/example/93976/sklearn.feature_selection.SelectFromModel

from matplotlib import pyplot as plt

feat_labels = data.columns[1:]

clf4.fit(X_train, y_train)
importances = clf4.feature_importances_

indices = np.argsort(importances)[::-1]

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
plt.title('Feature Importance')
plt.bar(range(X_train.shape[1]), importances[indices], align='center')

plt.xticks(range(X_train.shape[1]),
    feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

```python
# Assistance from DataCamp and https://www.programcreek.com/python/example/93976/sklearn.feature_selection.SelectFromModel

from sklearn.feature_selection import SelectFromModel

sfm = SelectFromModel(clf4, threshold=0.05, prefit=True)
X_selected = sfm.transform(X_train)
print('Number of features that meet this threshold criterion:', X_selected.shape[1])

for f in range(X_selected.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
```
